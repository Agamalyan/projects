### Шаг 1. Откройте файлы с данными и объедините их в один датафрейм.

Объедините данные таким образом, чтобы все объекты из датасета `mkrf_movies` обязательно вошли в получившийся датафрейм.

```python
import pandas as pd
```

```python
movies = pd.read_csv('/datasets/mkrf_movies.csv')
display(movies)
movies.info()
```

```python
shows = pd.read_csv('/datasets/mkrf_shows.csv')
pd.options.display.float_format = '{:,.2f}'.format
display(shows)
shows.info()
```

```python
movies['puNumber'].sort_values().unique()
```

В данных присутствуют нестандартные значения. Имеет смысл проводить объединение по строковому формату.

```python
shows['puNumber'] = shows['puNumber'].astype('str')
```

```python
movies = movies.merge(shows, how = 'left', on = 'puNumber')
```

**Проверим полученные данные и убедимся, что ничего не пропало после объединения.**

```python
movies.info()
```

**Судя по полученному результату, все прошло нормально**

### Шаг 2. Предобработка данных

#### Шаг 2.1. Проверьте типы данных

- Проверьте типы данных в датафрейме и преобразуйте их там, где это необходимо.

```python
movies.info()
```

В некоторых колонках данные неподходящего формата, следует изменить типы данных.

```python
movies['show_start_date'] = pd.to_datetime(movies['show_start_date'], format = ('%Y-%m-%dT%H:%M:%S'))
movies['refundable_support'] = movies['refundable_support'].astype('float')
```

В колонке 'ratings' вместе с дробными значениями присутствуют процентные значения. Так как маловероятно, чтобы у фильмов были оценки около 9-10, то данные значения стоит удалить, а остальныые значения перевести в числовой формат.

```python
movies['ratings'] = movies['ratings'].astype('str')
movies['ratings'].unique()
```

```python
movies['ratings'] = pd.to_numeric(movies['ratings'], errors = 'coerce')
```

Таким же способом следует изменить формат в остальных колонках.

```python
movies['puNumber'] = pd.to_numeric(movies['puNumber'],errors = 'coerce')
```

#### Шаг 2.2. Изучите пропуски в датафрейме

- Объясните, почему заполнили пропуски определённым образом или почему не стали этого делать.

Судя по информации о датафрейме в некоторых колонках присутствуют пропуски, следует изучить их детальнее.

```python
movies.info()
```

Пропуски имеются в следующих колонках: \
puNumber - можно оставить как есть \
film_studio - данный пропуск придется либо заполнять вручную через поисковик, либо оставить как есть, ведь существуют фильмы, снятые без помощи студий. \
production_country - заполнять также не обязательно, всего 2 пропуска и не влияет на анализ. \
director - искать придется также вручную в открытых источниках по каждому фильму.\
producer - ручной поиск\
refundable_support, nonrefundable_support, budget - так как информацию предоставило министерство культуры, их мало интересовал бюджет фильмов, которые они не финансировали, да и достоверной информации они не могут иметь, сравнивать голивудские фильмы с российскими также не имеет смысла, разница в бюджетах колоссальная, поэтому данные пропуски следует оставить как есть. Все равно в дальнейшем анализе они не помогут.\
ratings - также не стоит трогать, так как рейтинг присуждается на кинопоиске известным фильмам, если рейтинг отсутсвует, то либо мы сами его удалили из-за неверных значений в нем, либо фильм никого не заинтересовал и оценивать было некому.\
genres - также придется заполнять либо вручную, либо оставлять как есть, ведь для присуждения жанра фильм нужно просмотреть, а для 1000 значений это неуместно. Если в дальнейшем данная информация потребуется для анализа можно будет заняться их заполнением.\
box_office - общие сборы фильма в прокате. На самом деле такую информацию не все компании публикуют, так как придется работать с огромным массивом данных. Так как нас интересует российское кино, получившее госсударственную поддержку, то 3000 значений будет достаточно для анализа.
